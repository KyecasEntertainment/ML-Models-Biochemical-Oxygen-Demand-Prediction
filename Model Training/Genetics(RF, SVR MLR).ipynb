```python
import random
import numpy as np
import pandas as pd
from deap import base, creator, tools, algorithms
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
import pickle

# Assuming X_tikob, X_gunao, y_tikob, y_gunao are already defined
X_combined = pd.concat([X_tikob, X_gunao], axis=0)
y_combined = pd.concat([y_tikob, y_gunao], axis=0)

X_train, X_test, y_train, y_test = train_test_split(X_combined, y_combined, test_size=0.2, random_state=1)

def evaluate(individual):
    n_estimators = max(10, int(individual[0]))
    max_depth = max(1, int(individual[1]))
    min_samples_split = max(2, int(individual[2]))
    min_samples_leaf = max(1, int(individual[3]))
    
    rf = RandomForestRegressor(
        n_estimators=n_estimators,
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        min_samples_leaf=min_samples_leaf,
        random_state=42
    )
    
    rf.fit(X_train, y_train)
    predictions = rf.predict(X_test)
    mse = mean_squared_error(y_test, predictions)
    return mse,

# Check if the classes are already created
if not hasattr(creator, "FitnessMin"):
    creator.create("FitnessMin", base.Fitness, weights=(-1.0,))
if not hasattr(creator, "Individual"):
    creator.create("Individual", list, fitness=creator.FitnessMin)

toolbox = base.Toolbox()
toolbox.register("attr_int_n_estimators", random.randint, 10, 300)
toolbox.register("attr_int_max_depth", random.randint, 1, 30)
toolbox.register("attr_int_min_samples_split", random.randint, 2, 40)
toolbox.register("attr_int_min_samples_leaf", random.randint, 1, 20)

toolbox.register("individual", tools.initCycle, creator.Individual, 
                 (toolbox.attr_int_n_estimators, toolbox.attr_int_max_depth, toolbox.attr_int_min_samples_split, toolbox.attr_int_min_samples_leaf), n=1)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

toolbox.register("mate", tools.cxBlend, alpha=0.5)
toolbox.register("mutate", tools.mutGaussian, mu=0, sigma=1, indpb=0.1)
toolbox.register("select", tools.selTournament, tournsize=3)
toolbox.register("evaluate", evaluate)

# Run the genetic algorithm
population = toolbox.population(n=100)
NGEN = 20
CXPB = 0.7
MUTPB = 0.3

print("Training")

for gen in range(NGEN):
    offspring = algorithms.varAnd(population, toolbox, cxpb=CXPB, mutpb=MUTPB)
    fits = map(toolbox.evaluate, offspring)
    
    for fit, ind in zip(fits, offspring):
        ind.fitness.values = fit
    
    population = toolbox.select(offspring, k=len(population))
    
    best_individual = tools.selBest(population, k=1)[0]
    best_mse = best_individual.fitness.values[0]
    
    # Print progress
    print(f"Generation {gen+1}/{NGEN}:")
    print(f"  Best Individual: {best_individual}")
    print(f"  Best MSE: {best_mse}")
    
print('Training Complete')

# Train final model with best hyperparameters
best_rf = RandomForestRegressor(
    n_estimators=max(10, int(best_individual[0])),
    max_depth=max(1, int(best_individual[1])),
    min_samples_split=max(2, int(best_individual[2])),
    min_samples_leaf=max(1, int(best_individual[3])),
    random_state=42
)

best_rf.fit(X_train, y_train)
final_predictions = best_rf.predict(X_test)
final_mse = mean_squared_error(y_test, final_predictions)
final_rmse = np.sqrt(final_mse)
final_mae = mean_absolute_error(y_test, final_predictions)
final_r2 = r2_score(y_test, final_predictions)
final_mape = np.mean(np.abs((y_test - final_predictions) / y_test)) * 100

print('Final Test MSE:', final_mse)
print('Final Test RMSE:', final_rmse)
print('Final Test MAE:', final_mae)
print('Final Test R^2:', final_r2)
print('Final Test MAPE:', final_mape)

# Save the trained model to a file
with open('best_random_forest_model.pkl', 'wb') as model_file:
    pickle.dump(best_rf, model_file)
print('Model saved to best_random_forest_model.pkl')

# Load the model from a file (example usage)
with open('best_random_forest_model.pkl', 'rb') as model_file:
    loaded_model = pickle.load(model_file)
print('Model loaded from best_random_forest_model.pkl')

# You can now use the loaded model to make predictions
loaded_predictions = loaded_model.predict(X_test)
loaded_final_mse = mean_squared_error(y_test, loaded_predictions)
loaded_final_rmse = np.sqrt(loaded_final_mse)
loaded_final_mae = mean_absolute_error(y_test, loaded_predictions)
loaded_final_r2 = r2_score(y_test, loaded_predictions)
loaded_final_mape = np.mean(np.abs((y_test - loaded_predictions) / y_test)) * 100

print('Loaded Model Test MSE:', loaded_final_mse)
print('Loaded Model Test RMSE:', loaded_final_rmse)
print('Loaded Model Test MAE:', loaded_final_mae)
print('Loaded Model Test R^2:', loaded_final_r2)
print('Loaded Model Test MAPE:', loaded_final_mape)
```

```python
import random
import numpy as np
import pickle
from deap import base, creator, tools, algorithms
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split, cross_val_score, KFold, RepeatedKFold

# Example data, replace with your actual data
X = np.random.rand(120, 10)
y = np.random.rand(120)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.009, random_state=42)

# Normalize your data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Evaluate function
def evaluate(individual):
    degree = int(individual[0] * 8) + 1  # Polynomial degree between 1 and 8
    include_bias = bool(int(individual[1]))  # Include bias in polynomial features
    model_choice = int(individual[2] * 4)  # Choose between LinearRegression, Ridge, Lasso, ElasticNet
    alpha = individual[3] * 10  # Regularization parameter between 0 and 10
    
    poly = PolynomialFeatures(degree=degree, include_bias=include_bias)
    X_train_poly = poly.fit_transform(X_train_scaled)
    X_test_poly = poly.transform(X_test_scaled)
    
    if model_choice == 0:
        model = LinearRegression()
    elif model_choice == 1:
        model = Ridge(alpha=alpha)
    elif model_choice == 2:
        model = Lasso(alpha=alpha)
    else:
        model = ElasticNet(alpha=alpha)
    
    kf = RepeatedKFold(n_splits=5, n_repeats=2, random_state=42)
    scores = -cross_val_score(model, X_train_poly, y_train, cv=kf, scoring='neg_mean_squared_error')
    
    mse = np.mean(scores)
    return mse,

# Create fitness and individual classes
if not hasattr(creator, "FitnessMin"):
    creator.create("FitnessMin", base.Fitness, weights=(-1.0,))
if not hasattr(creator, "Individual"):
    creator.create("Individual", list, fitness=creator.FitnessMin)

# Register toolbox functions
toolbox = base.Toolbox()
toolbox.register("attr_float", random.random)
toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_float, n=4)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

toolbox.register("mate", tools.cxTwoPoint)
toolbox.register("mutate", tools.mutFlipBit, indpb=0.05)
toolbox.register("select", tools.selTournament, tournsize=3)
toolbox.register("evaluate", evaluate)

# Run the genetic algorithm
population = toolbox.population(n=1)  # Increased population size
NGEN = 1  # Increased number of generations
CXPB = 0.7  # Crossover probability
MUTPB = 0.2  # Mutation probability

print("Training")

for gen in range(NGEN):
    offspring = algorithms.varAnd(population, toolbox, cxpb=CXPB, mutpb=MUTPB)
    fits = map(toolbox.evaluate, offspring)
    
    for fit, ind in zip(fits, offspring):
        ind.fitness.values = fit
    
    population = toolbox.select(offspring, k=len(population))
    
    best_individual = tools.selBest(population, k=1)[0]
    best_mse = best_individual.fitness.values[0]
    
    # Print progress
    if gen % 10 == 0 or gen == NGEN - 1:
        print(f"Generation {gen+1}/{NGEN}:")
        print(f"  Best Individual: {best_individual}")
        print(f"  Best MSE: {best_mse}")
    
print('Training Complete')

# Train final model with the best hyperparameters
best_degree = int(best_individual[0] * 8) + 1
best_include_bias = bool(int(best_individual[1]))
best_model_choice = int(best_individual[2] * 4)
best_alpha = best_individual[3] * 10

poly = PolynomialFeatures(degree=best_degree, include_bias=best_include_bias)
X_train_poly = poly.fit_transform(X_train_scaled)
X_test_poly = poly.transform(X_test_scaled)

if best_model_choice == 0:
    best_model = LinearRegression()
elif best_model_choice == 1:
    best_model = Ridge(alpha=best_alpha)
elif best_model_choice == 2:
    best_model = Lasso(alpha=best_alpha)
else:
    best_model = ElasticNet(alpha=best_alpha)

best_model.fit(X_train_poly, y_train)
final_predictions = best_model.predict(X_test_poly)
final_mse = mean_squared_error(y_test, final_predictions)
final_rmse = np.sqrt(final_mse)
final_mae = mean_absolute_error(y_test, final_predictions)
final_r2 = r2_score(y_test, final_predictions)
final_mape = np.mean(np.abs((y_test - final_predictions) / y_test)) * 100

print('Final Test MSE:', final_mse)
print('Final Test RMSE:', final_rmse)
print('Final Test MAE:', final_mae)
print('Final Test R^2:', final_r2)
print('Final Test MAPE:', final_mape)

# Save the trained model and polynomial feature transformer using pickle
with open('best_model.pkl', 'wb') as model_file:
    pickle.dump(best_model, model_file)
with open('poly_transformer.pkl', 'wb') as transformer_file:
    pickle.dump(poly, transformer_file)

print('Model and transformer saved to pickle files.')

# Example of loading and using the model and transformer
with open('best_model.pkl', 'rb') as model_file:
    loaded_model = pickle.load(model_file)
with open('poly_transformer.pkl', 'rb') as transformer_file:
    loaded_poly = pickle.load(transformer_file)

X_test_poly_loaded = loaded_poly.transform(X_test_scaled)
loaded_predictions = loaded_model.predict(X_test_poly_loaded)

loaded_mse = mean_squared_error(y_test, loaded_predictions)
loaded_rmse = np.sqrt(loaded_mse)
loaded_mae = mean_absolute_error(y_test, loaded_predictions)
loaded_r2 = r2_score(y_test, loaded_predictions) 
loaded_mape = np.mean(np.abs((y_test - loaded_predictions) / y_test)) * 100

print('Loaded Model Test MSE:', loaded_mse)
print('Loaded Model Test RMSE:', loaded_rmse)
print('Loaded Model Test MAE:', loaded_mae)
print('Loaded Model Test R^2:', loaded_r2)
print('Loaded Model Test MAPE:', loaded_mape)
```
